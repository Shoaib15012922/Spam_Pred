```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
data = pd.read_csv("Spam.csv")
```

```python
data.head()
```

```python
Xtrain = data.drop('spam',axis  = 1)
ytrain = data[['spam']] 
Xtrain_with_intercept = sm.add_constant(Xtrain)
   

spam_fit = sm.Logit(ytrain, Xtrain_with_intercept).fit()

print(spam_fit.summary())
```

```python
predictions = spam_fit.predict(data)
```
